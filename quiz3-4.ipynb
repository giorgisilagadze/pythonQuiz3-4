```python
#replace ფუნქციით შევცვალეთ Earthquake ქართული შესატყვისით
```

```python
df["Type"] = df["Type"].str.replace("Earthquake", "miwisdzvra")
```

```python
df
```

```python
# მიწისძვრის სიღრმის მიხედვით შევქმნათ 3 კარეგორია და cut ფუნქციით დავყოთ
```

```python

bins = [15.00, 79.00, 80.00, 131.60]
labels = ["small Depth", "medium Depth", "strong Depth"]

df['DepthTypes'] = pd.cut(df['Depth'], bins=bins, labels=labels, right=False)
df
```

```python

import pandas as pd
import matplotlib.pyplot as plt
```

```python
#box-plot მაგნიტუდის მიხედვით
plt.boxplot(df['Magnitude'], vert=False)
plt.title('Box Plot of Magnitude')
plt.xlabel('Magnitude')
plt.show()
```

```python
# ხაზოვანი დიაგრამა
count = df['Magnitude Type'].value_counts()

plt.figure(figsize=(10, 6))
count.plot(kind='line', marker='o', color='b')
plt.title('ხაზოვანი დიაგრამა მაგნიტუდის ტიპისთვის')
plt.xlabel('Magnitude Type')
plt.ylabel('სიხშირე')
plt.grid(True)
plt.show()
```

```python
# ჰისტოგრამა
```

```python
plt.hist(df['Depth'], bins=30, color='red', edgecolor='green')
plt.title('სიღმეების ჰისტოგრამა')
plt.xlabel('Depth')
plt.ylabel('სიხშირე')
plt.show()
```

```python
#წრიული დიაგრამა
```

```python
magnitude_type_counts = df['Magnitude Type'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(magnitude_type_counts, labels=magnitude_type_counts.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('წრიული დიაგრამა')
plt.show()
```
